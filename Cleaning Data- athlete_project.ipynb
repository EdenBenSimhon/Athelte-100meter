# Data Cleaning 

```python
import pandas as pd
import time 
import os       
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from datetime import datetime, date
```

```python
df_rec=pd.read_csv("step 1-records after crawling.csv")
```

```python
copy_df = df_rec.copy()
```

```python
copy_df
```

Delete irrelevant columns 

```python
 copy_df=copy_df.drop(columns=["Unnamed: 0"])

```

```python
copy_df
```

```python
copy_df.info()
```

Type change in Mark Result Score columns.
In addition, cleaning is performed in the Mark column

```python
copy_df['Mark'] = copy_df['Mark'].str.replace('h', '')
copy_df['Mark'] = copy_df['Mark'].astype(float)
```

```python
copy_df["Result Score"]=copy_df["Result Score"].astype(int)
```

```python
copy_df.info()
```

Cleaning the Wind column and changing the type.

```python
copy_df['Wind'] = copy_df['Wind'].str.replace('w',"")
copy_df['Wind']=pd.to_numeric(copy_df['Wind'].str.replace(',',''), errors='coerce')
```

```python
copy_df['Wind']
```

sort the table by the Result Score column

```python
copy_df=copy_df.sort_values(by='Result Score', ascending=False)
```

```python
copy_df.info()
```

```python
copy_df
```

```python
copy_df['Date'] = copy_df['Date'].str.replace('JAN', '01')
copy_df['Date'] = copy_df['Date'].str.replace('FEB', '02')
copy_df['Date'] = copy_df['Date'].str.replace('MAR', '03')
copy_df['Date'] = copy_df['Date'].str.replace('APR', '04')
copy_df['Date'] = copy_df['Date'].str.replace('MAY', '05')
copy_df['Date'] = copy_df['Date'].str.replace('JUN', '06')
copy_df['Date'] = copy_df['Date'].str.replace('JUL', '07')
copy_df['Date'] = copy_df['Date'].str.replace('AUG', '08')
copy_df['Date'] = copy_df['Date'].str.replace('SEP', '09')
copy_df['Date'] = copy_df['Date'].str.replace('OCT', '10')
copy_df['Date'] = copy_df['Date'].str.replace('NOV', '11')
copy_df['Date'] = copy_df['Date'].str.replace('DEC', '12')
copy_df['Date'] = copy_df['Date'].str.replace(' ', '-')
copy_df=copy_df.sort_values(by='Age', ascending=False)


```

```python
copy_df
```

```python
copy_df.info()
```

Change of birth dates to age

```python
copy_df["Age"] = pd.to_datetime(copy_df["Age"])

copy_df["Age"] = copy_df["Age"].apply(lambda x : (pd.datetime.now().year - x.year))
copy_df
```

Adding Age of competition column.

```python
Ageofcomp=list()

for i in copy_df["Date"]:
    Ageofcomp.append((i[-4:]))
Ageofcomp
```

```python
yearDOB=list()
result=list()
for i in copy_df['Age']:
        yearDOB.append((2021-i))

copy_df['Age of competition']=    Ageofcomp
copy_df['Age of competition'] = copy_df['Age of competition'].astype(int)
```

```python
copy_df['Age of competition']=copy_df['Age of competition']-yearDOB
```
