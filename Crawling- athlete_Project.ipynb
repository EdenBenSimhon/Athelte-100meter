# top list records all Time -100 meters

```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd


Rank=list()
Competitors=list()
DOBs=[]
Countries=list()
Scores=list()
Event_Lists=list()
Gender=list()
```

Create list to add elements 

```python
rank=list()
Mark=list()
Wind=list()
Competitors=list()
DoB=list()
country=list()
Position=list()
Venue=list()
date_event=list()
resultscore=list()
sex=list()
```

Performing a crawl on breaking records at all times- 100 meter Men

```python
rec=1
pages_rec=296
for rec in range (1,pages_rec):
    rec = str(rec)
    toplist="https://www.worldathletics.org/records/all-time-toplists/sprints/100-metres/outdoor/men/senior?regionType=world&timing=all&windReading=all&page="+rec+"&bestResultsOnly=false&firstDay=1900-01-01&lastDay=2021-05-15"
    html=requests.get(toplist)
    soup=BeautifulSoup(html.content,"html.parser")
    table=soup("table",attrs={"class":"records-table"})[0]
    for row in table.find_all("tr"):
        cells=row.find_all("td")
        if(len(cells)<10):
            continue
        rank.append(cells[0].get_text().strip())
        Mark.append(cells[1].get_text().strip())
        Wind.append(cells[2].get_text().strip())
        Competitors.append(cells[3].get_text().strip())
        DoB.append(cells[4].get_text().strip())
        country.append(cells[5].get_text().strip())
        Position.append(cells[6].get_text().strip())
        Venue.append(cells[8].get_text().strip())
        date_event.append(cells[9].get_text().strip())
        resultscore.append(cells[10].get_text().strip())
        sex.append("Male")

```

Performing a crawl on breaking records at all times- 100 meter Women

Create Dataframe

```python
Df_rec_100meter=pd.DataFrame({"Rank":rank,"Mark":Mark,"Wind":Wind,"Competitor":Competitors,"Gender":sex,"Age":DoB,"Country":country,"Position":Position,"Venue":Venue,"Date":date_event,"Result Score":resultscore})
```
