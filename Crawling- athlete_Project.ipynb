```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd


Rank=list()
Competitors=list()
DOBs=[]
Countries=list()
Scores=list()
Event_Lists=list()
Gender=list()
```

# Ranking world athlete

Performing a crawl on ranking world athete

```python
page=1
pages=121
for page in range (1,pages):
    page = str(page)
    url="https://www.worldathletics.org/world-rankings/overall-ranking/men?regionType=world&page="+page+"&rankDate=2021-05-11&limitByCountry=0"
    html=requests.get(url)
    soup=BeautifulSoup(html.content,"html.parser")
    table=soup("table",attrs={"class":"records-table"})[0]
    for row in table.find_all("tr"):
        cells=row.find_all("td")
        if(len(cells)<6):
            continue
        Rank.append(cells[0].get_text().strip())
        Competitors.append(cells[1].get_text().strip())
        DOBs.append(cells[2].get_text().strip())
        Countries.append(cells[3].get_text().strip())
        Scores.append(cells[4].get_text().strip())
        Event_Lists.append(cells[5].get_text().strip())
        Gender.append("Male")

```

```python
page=1
pages=111
for page in range (1,pages):
    page = str(page)
    url="https://www.worldathletics.org/world-rankings/overall-ranking/women?regionType=world&page="+page+"&rankDate=2021-05-11&limitByCountry=0"
    html=requests.get(url)
    soup=BeautifulSoup(html.content,"html.parser")
    table=soup("table",attrs={"class":"records-table"})[0]
    for row in table.find_all("tr"):
        cells=row.find_all("td")
        if(len(cells)<6):
            continue
        Rank.append(cells[0].get_text().strip())
        Competitors.append(cells[1].get_text().strip())
        DOBs.append(cells[2].get_text().strip())
        Countries.append(cells[3].get_text().strip())
        Scores.append(cells[4].get_text().strip())
        Event_Lists.append(cells[5].get_text().strip())
        Gender.append("Female")
```

```python
df=pd.DataFrame({"Place":Rank,"Competitor":Competitors,"Gender":Gender,"Date of birth":DOBs,"Contry":Countries,"score":Scores,"Event Lists":Event_Lists})
```

```python
df.to_csv('out.csv')
```

# top list records all Time

Performing a crawl on breaking records at all times- 100 meter Men

```python
rank=list()
Mark=list()
Wind=list()
Competitors=list()
DoB=list()
country=list()
Position=list()
Venue=list()
date_event=list()
resultscore=list()
sex=list()
```

```python
rec=1
pages_rec=296
for rec in range (1,pages_rec):
    rec = str(rec)
    toplist="https://www.worldathletics.org/records/all-time-toplists/sprints/100-metres/outdoor/men/senior?regionType=world&timing=all&windReading=all&page="+rec+"&bestResultsOnly=false&firstDay=1900-01-01&lastDay=2021-05-15"
    html=requests.get(toplist)
    soup=BeautifulSoup(html.content,"html.parser")
    table=soup("table",attrs={"class":"records-table"})[0]
    for row in table.find_all("tr"):
        cells=row.find_all("td")
        if(len(cells)<10):
            continue
        rank.append(cells[0].get_text().strip())
        Mark.append(cells[1].get_text().strip())
        Wind.append(cells[2].get_text().strip())
        Competitors.append(cells[3].get_text().strip())
        DoB.append(cells[4].get_text().strip())
        country.append(cells[5].get_text().strip())
        Position.append(cells[6].get_text().strip())
        Venue.append(cells[8].get_text().strip())
        date_event.append(cells[9].get_text().strip())
        resultscore.append(cells[10].get_text().strip())
        sex.append("Male")

```
