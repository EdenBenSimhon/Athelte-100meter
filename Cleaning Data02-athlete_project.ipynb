# Data Cleaning-02


keep cleaning the data 

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("step 2-Cleaning Data records.csv")
```

```python
df=df.drop(columns=["Unnamed: 0"])
df
```

```python
df.info()
```

Create a column that shows the speed per meter

```python
df["mps"]=100/df['Mark']
```

```python
df["mps"]
```

```python
df.sort_values(by='mps', ascending=False)
df['Position']
```

Position column cleaning and converts types to int

```python
df['Position']
tuple_list= df['Position']
position=list()
tuple_list=tuple_list.astype(str)
for i in tuple_list:
    position.append(i[0])

df['Position']=position
```

```python
df['Position']
```

```python
df['Position']  = df['Position'].str.replace('h', '')
df['Position']  = df['Position'].str.replace('s', '')
df['Position']  = df['Position'].str.replace('q', '')
df['Position']  = df['Position'].str.replace('f', '')
df['Position']  = df['Position'].str.replace('n', '')
df['Position']  = df['Position'].str.replace('c', '')
```

```python
df['Position']=pd.to_numeric(df['Position'], errors='coerce').convert_dtypes() 
```

Changing Gender column to numric

```python
replace_map={'Female':2,'Male':1}
df.replace(replace_map,inplace=True)
```

```python
df=df.sort_values(by='Rank', ascending=True)
```

```python
df['Gender']
```

Creating new column Kilometer per hours

```python
df["kmh"]=0.1/(df["Mark"]/3600)
```

```python
df
```

Deleting duplicates Competitor

```python
df=df.drop_duplicates(subset=["Competitor"])
df
```

```python
df.info()
```

```python
df.to_csv('step 3-Cleaning Data records.csv')
```
